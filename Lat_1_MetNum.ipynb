```python
import matplotlib.pyplot as plt
import numpy as np

def factorial(n):
  f = 1
  if n == 0:
      return 1
  elif n >= 2:
    for k in range(2, 100):
          f = f*k
    return f

def exp(x, garis):
  s = 0
  for i in range(2, garis):
        s = s + (x**i)/factorial(i)
  return s

x = np.arange(0,15,0.001)
plt.plot(x,exp(x, 100), "red")
plt.plot(x,exp(x, 10), "blue")
plt.plot(x,exp(x, 5), "green")
plt.show()
```
